```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/version 1.csv")
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].str.lower().str.strip()
```

```python
df.head()
```

```python
df['Gender'].replace({'m': 'Male', 'f': 'Female'}, inplace=True)
```

```python
df.head()
```

```python
#Identified the catergotical colums for standization
#categorical_colums =cat_cols
cat_cols = ["Gender", "Location", "Education", "Occupation", "Primary_App", 
                    "Secondary_Apps", "Usage_Frequency", "Reason_for_Using", 
                    "Challenges", "Desired_Features", "Preferred_Communication"]
for col in cat_cols:
    df[col] = df[col].astype(str).str.lower().str.strip()
```

```python
# Check for Missing Values
missing_values = df.isnull().sum()
print("Missing Values per Column:\n", missing_values)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)

```

```python
df['Gender'] = df['Gender'].isnull().astype(int)
```

```python
df.head()
```

```python
 #Converted 'Daily_Usage_Time' to Numeric (Minutes)
time_mapping = {
    '30 minutes': 30,
    '1 hour': 60,
    '1.5 hours': 90,
    '2 hours': 120,
    '3 hours': 180
}
df['Daily_Usage_Time'] = df['Daily_Usage_Time'].map(time_mapping)
```

```python
df[df.select_dtypes(include='object').columns] = df.select_dtypes(include='object').astype('category')
```

```python
df.info()
```

```python
# Detecting the Outliers using IQR Method
numerical_columns = ['Age', 'Satisfaction', 'Daily_Usage_Time']
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Outliers in {col}: {len(outliers)}")
```

```python
#  A Boxplot for The Outlier Visualization
plt.figure(figsize=(15, 10))
sns.boxplot(data=df[numerical_columns])
plt.xticks(rotation=90)
plt.show()
```

```python
# Save Cleaned Data (Optional)
df.to_csv("../data/version 1.csv", index=False)
```
